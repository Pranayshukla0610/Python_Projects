```python
#Read a CSV

import pandas as pd

df = pd.read_csv('/content/creditcard.csv')

print(df.head(10))
print(df.isnull().sum())
print(df.dtypes)
print(df.duplicated().sum())
```

```python
#Read a 5GB CSV without crashing memory

chunks = pd.read_csv('/content/creditcard.csv',chunksize=100000)

for chunk in chunks:
  print(chunk.shape)
```

```python
#Read CSV faster using Polars

import polars as pl

df = pl.read_csv(
    '/content/creditcard.csv',
    schema_overrides={
        "Time": pl.Float64
    }
)

print(df.head())
print(df.schema)
```

```python
#Read SQL table into pandas

from sqlalchemy import create_engine
import pandas as pd

engine = create_engine(
    "postgresql://user:password@localhost:5432/mydb"
)

df = pd.read_sql("SELECT * FROM employees", engine)

print(df.head())
```

```python
#Read nested JSON

import pandas as pd

df = pd.read_json('users.json')
print(df.head())
```

```python
#Flatten nested JSON

import pandas as pd
import json

with open('nested.json') as f:
  data = json.load(f)

df = pd.json_normalize(data)
print(df.head())
```

```python
import xml.etree.ElementTree as ET
tree = ET.parse('employee.xml')

root = tree.getroot()

for emp in root.findall('employee'):
  print(emp.find("name").text)
```
