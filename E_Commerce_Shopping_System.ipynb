```python
from abc import ABC, abstractmethod

class Product:
  product_count = 0

  def __init__(self,name,price):
    self.name = name
    self.__price = price
    Product.product_count += 1

  def get_price(self):
    return self.__price

  def __str__(self):
    return f"{self.name} - ${self.__price}"

  @classmethod
  def total_products(cls):
    return cls.product_count

  @staticmethod
  def validate_price(price):
    return price > 0
```

```python
#CART CLASS

class Cart:
  def __init__(self):
    self.items = []

  def add_item(self,product):
    self.items.append(product)

  def remove_product(self,name):
    self.items = [item for item in self.items if item.name != name]

  def total(self):
    return sum(item.get_price() for item in self.items)

  def show_items(self):
    for item in self.items:
      print(item)

  def __len__(self):
    return len(self.items)
```

```python
#USER CLASS
class User:
  def __init__(self,name):
    self.name = name
    self.cart = Cart()

  def add_to_cart(self,product):
    self.cart.add_item(product)

  def remove_from_cart(self,name):
    self.cart.remove_product(name)
```

```python
#ORDER CLASS

class Order:
  def __init__(self, user, items):
    self.user = user
    self.items = items

  def total_price(self):
    return sum(item.get_price() for item in self.items)
```

```python
#PAYMENT ABSTRACT CLASS
class Payment(ABC):
  @abstractmethod
  def process(self,order):
    pass
```

```python
#UPI PAYMENT
class UPI(Payment):
  def process_payment(self,amount):
    print(f"Processing UPI payment of ${amount}")
```

```python
#Credit Card Pyament
class CreditCard(Payment):
  def process_payment(self,amount):
    print(f"Processing Credit Card Payment of ${amount}")
```

```python
#STORE MANAGER CLASS
class Store:
  def __init__(self):
    self.products = []
    self.orders = []

  def add_product(self,product):
    self.products.append(product)

  def show_product(self):
    for p in self.products:
      print(p)

  def final_product(self, name):
    for p in self.products:
      if p.name == name:
        return p
    return None

  def place_order(self,user,payment_method):
    order = Order(user, user.cart.items)
    amount = order.total_price()
    payment_method.process_payment(amount)
    self.orders.append(amount)
    user.cart.items = []
    print("Order Placed Successfully")
```

```python
#STORE SYSTEM (CLI)

class StoreSystem:
  def __init__(self):
    self.store = Store()
    self.users = {}

  def add_user(self,name):
    self.users[name] = User(name)

  def get_user(self,name):
    return self.users.get(name)

  def start(self):

    while True:
      print("\n1 Show Product")
      print("2 Add to Cart")
      print("3 Show Cart")
      print("4 Place Order")
      print("5 Exit")

      choice = input("Enter your choice: ")

      if choice == "1":
        self.store.show_product()

      elif choice == "2":
        user = input("User Name:")
        product = input("Product Name:")

        u = self.get_user(user)
        p = self.store.find_product(product)

      if u and p:
        u.add_to_cart(p)

      elif choice == "3":
        user = input("User Name:")
        u = self.get_user(user)

        if u:
          u.cart.show_items()

      elif choice == "4":
        user = input("User Name:")
        payment = input("Payment (UPI/card):")

      u = self.get_user()
      if payment == "UPI":
        payment_method = UPI()
      elif payment == "card":
        payment_method = CreditCard()

        self.store.place_order(u, pay)

      elif choice == "5":
        break
```

```python
# RUN PROGRAM
system = StoreSystem()

system.store.add_product(Product("Laptop", 1000))
system.store.add_product(Product("Phone", 500))
system.store.add_product(Product("Headphones", 100))

system.add_user("Alice")

system.start()
```
