```python
#BOOK CLASS

from abc import ABC, abstractmethod

class Book:
  total_books = 0

  def __init__(self, title, author):
    self.title = title
    self.author = author
    self.__is_borrowed = False
    Book.total_books += 1

  def borrow(self):
    if not self.__is_borrowed:
      self.__is_borrowed = True
      return True
    return False

  def return_book(self):
    self.__is_borrowed = False

  def is_available(self):
    return not self.__is_borrowed

  def __str__(self):
    status = "Available" if self.__is_available() else "Borrowed"
    return f"{self.title} by {self.author} [{status}]"

  @classmethod
  def get_total_books(cls):
    return cls.total_books

  @staticmethod
  def validate_title(title):
    return len(title) > 0
```

```python
#Abstract User Class

class User(ABC):
  def __init__(self,name):
    self.name = name
    self.borrowed_books = []

  @abstractmethod
  def borrow_book(self,library,title):
    pass

  def return_book(self,library, title):
    library.return_book(title,self)
```

```python
#Student Class

class Student(User):
  MAX_LIMIT = 3

  def borrow_book(self,library,title):
    if len(self.borrowed_books) >= Student.MAX_LIMIT:
      print("Borrow Limit Reached")

    book = library.borrow_book(title)

    if book:
      self.borrowed_books.append(book)
      print(self.name,"borrowed",title)
```

```python
#Librarian Class
class Librarian(User):
  def borrow_book(self,library,title):
    book = library.borrow_book(title)
    if book:
      self.borrowed_books.append(book)

  def add_book(self, library, title, author):
    library.add_book(Book(title,author))
```

```python
#Library Class
class Library:
  def __init__(self):
    self.books = []

  def add_book(self,book):
    self.books.append(book)

  def borrow_book(self,title):
    for book in self.books:
      if book.title == title and book.is_available():
        book.borrow()
        return book
    print("Book not available")
    return None

  def return_book(self,title,user):
    for book in user.borrowed_books:
      if book.title == title:
        book.return_book()
        user.borrowed_books.remove(book)
        print("Book Returned")
        return

  def show_books(self):
    for book in self.books:
      print(book)


  def __len__(self):
    return len(self.books)
```

```python
#Library System (CLI)

class LibrarySystem:
  def __init__(self):
    self.library = Library()
    self.users = {}

  def add_user(self, name, user_type):
    if user_type == "student":
      self.users[name] = Student(name)
    else:
      self.users[name] = Librarian(name)

  def get_user(self,name):
    return self.users.get(name)

  def start(self):
    while True:
      print("\n1 Add Book")
      print("2 Show Book")
      print("3 Borrow Book")
      print("4 Return Book")
      print("5 Exit")

      choice = input("Choice:")

      if choice == "1":
        title = input("Title:")
        author = input("Author:")
        self.library.add_book(Book(title,author))

      elif choice == "2":
        self.library.show_books()

      elif choice == "3":
        name = input("UserName:")
        title = input("Title:")
        user = self.get_user(name)

        if user:
          user.borrow_book(self.library, title)

      elif choice == "4":
        name = input("UserName:")
        title = input("Book Title:")
        if user:
          user.return_book(self.library,title)

      elif choice == "5":
        break
```

```python
system = LibrarySystem()
system.add_user("Alice","student")
system.add_user("Bob","librarian")
system.start()
```
