```python
import time
import timeit
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numbers = list(range(1000000))
```

```python
#FOR Loop

def for_loop(numbers):
  result = []

  for i in numbers:
    result.append(i * 2)

  return result
```

```python
#WHILE Loop

def while_loop(numbers):
  result = []
  i = 0

  while i < len(numbers):
    result.append(numbers[i] * 2)
    i += 1

  return result
```

```python
#List Comprehension

def list_comprehension(numbers):
  return list(map(lambda x:x * 2, numbers))
```

```python
def numpy_vectorization(numbers):
  arr = np.array(numbers)
  return arr * 2
```

```python
#Loop Control Statements
#Break Example

for i in range(10):
  if i == 5:
    break

  print(i)
```

```python
#Continue Example

for i in range(10):
  if i == 5:
    continue
  print(i)
```

```python
#Pass Example

for i in range(5):
  if i == 3:
    pass
  print(i)
```

```python
#Measure Execution Time
#Using time Module

start = time.time()
for_loop(numbers)

end = time.time()

print("Execution Time:",end - start)
```

```python
def map_function(numbers):
    return list(map(lambda x: x * 2, numbers))
```

```python
results = {}

results['For Loop'] = timeit.timeit(
    lambda: for_loop(numbers),
    number=5
)

results['While Loop'] = timeit.timeit(
    lambda: while_loop(numbers),
    number=5
)

results['List Comprehension'] = timeit.timeit(
    lambda: list_comprehension(numbers),
    number=5
)

results['Map Function'] = timeit.timeit(
    lambda: map_function(numbers),
    number=5
)

results['NumPy Vectorization'] = timeit.timeit(
    lambda: numpy_vectorization(numbers),
    number=5
)

print(results)
```
