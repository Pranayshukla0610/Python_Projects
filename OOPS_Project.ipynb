```python
#🏦 1. Banking System
#Concepts:
#Encapsulation → balance protection
#Inheritance → Account types
#Polymorphism → different interest logic
#Abstraction → base Account class

from abc import ABC, abstractmethod

class Account(ABC):
  def __init__(self,owner,balance=0):
    self.owner = owner
    self.__balance = balance

  def deposit(self,amount):
    self.__balance += amount

  def get_balance(self):
    return self.__balance

  @abstractmethod
  def interest_rate(self):
    pass

class SavingAccount(Account):
  def interest_rate(self):
    return self.get_balance() * 0.04

class CurrentAccount:
  def interest_rate(self):
    return self.get_balance() * 0.02

acc1 = SavingAccount('Amit',1000)
acc2 = CurrentAccount()
print(acc1.get_balance())
print(acc1.interest_rate())
```

```python
#🛒 2. E-Commerce Cart System
#Concepts:
#Composition → Cart contains Items
#Encapsulation → price handling
#Polymorphism → discount strategies
#Inheritance → Product types

from abc import ABC, abstractmethod

class Product:
    def __init__(self, name, price):
        self.name = name
        self.__price = price  # encapsulation

    def get_price(self):
        return self.__price


class DiscountStrategy(ABC):
    @abstractmethod
    def apply_discount(self, total):
        pass


class NoDiscount(DiscountStrategy):
    def apply_discount(self, total):
        return total


class PercentageDiscount(DiscountStrategy):
    def __init__(self, percent):
        self.percent = percent

    def apply_discount(self, total):
        return total * (1 - self.percent / 100)


class Cart:
    def __init__(self, discount_strategy):
        self.items = []
        self.discount_strategy = discount_strategy  # composition

    def add_item(self, product):
        self.items.append(product)

    def total(self):
        return sum(item.get_price() for item in self.items)

    def final_total(self):
        return self.discount_strategy.apply_discount(self.total())


p1 = Product("Laptop", 50000)
p2 = Product("Mouse", 2000)

cart = Cart(PercentageDiscount(10))
cart.add_item(p1)
cart.add_item(p2)

print(cart.final_total())
```
