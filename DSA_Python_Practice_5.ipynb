```python
#Two sum

arr = list(map(int,input("Enter array elements:").split()))

target = int(input("Enter target:"))

lookup = {}

for i in range(len(arr)):
  complement = target - arr[i]

  if complement in lookup:
    print("Indices:",lookup[complement], i)
    break
  lookup[arr[i]] = i
else:
  print("No solution found")
```

```python
#Remove Duplicates from Sorted Array

arr = list(map(int, input("Enter digits: ")))

arr.sort()
seen = []

for i in range(len(arr)):
  if arr[i] not in seen:
    seen.append(arr[i])

print(seen)
```

```python
#Remove Duplicates from Sorted Array

arr = list(map(int, input("Enter digits:").split()))

i = 0

for j in range(1, len(arr)):
  if arr[i] != arr[j]:
    i += 1
    arr[i] = arr[j]

print("Length:", i + 1)
print("Array:", arr[:i + 1])
```

```python
#Plus One

arr = list(map(int, input("Enter digits").split()))

for i in range(len(arr)-1,-1,-1):
  if arr[i] < 9:
    arr[i] += 1
    print(arr)
    break

  arr[i] = 0
  if i == 0:
    arr.insert(0,1)
    print(arr)
```

```python
#Remove Element

arr = list(map(int, input("Enter digits:").split()))
val = int(input("Enetr Value to be removed:"))
i = 0

for j in range(len(arr)):
  if arr[j] != val:
    arr[i] = arr[j]
    i += 1

print("Length:",i)
print("Array:",arr[:i])
```
