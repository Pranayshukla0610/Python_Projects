```python
import time
import matplotlib.pyplot as plt
```

```python
states = ['RED','GREEN','YELLOW']
```

```python
transitions = {
    "RED":"GREEN",
    "GREEN":"YELLOW",
    "YELLOW":"RED"
}
```

```python
class TrafficLightStateMachine:

    def __init__(self):
        self.current_state = "RED"

    def transition(self):

        print(f"Current State: {self.current_state}")

        self.current_state = transitions[self.current_state]

        print(f"Next State: {self.current_state}")
        print("-" * 30)

    def run(self, cycles):

        for _ in range(cycles):

            self.transition()

            time.sleep(1)
```

```python
machine = TrafficLightStateMachine()

machine.run(6)
```

```python
class AdvanceTrafficLight:
  def __init__(self):
    self.state='RED'

  def handle_event(self,event):
    if self.state=='RED' and event=='timer':
      self.state='GREEN'

    elif self.state=='GREEN' and event == 'timer':
      self.state='YELLOW'

    elif self.state=='YELLOW' and event == 'timer':
      self.state='RED'

    else:
      print('Invalid Transition')

    print("Current State:",self.state)
```

```python
traffic = AdvanceTrafficLight()

events = ['timer','timer','timer','timer']

for event in events:
  traffic.handle_event(event)
```

```python
class SmartTrafficLight:

    def __init__(self):

        self.state = "RED"
        self.history = []

    def transition(self):

        self.history.append(self.state)

        self.state = transitions[self.state]

    def run(self, cycles):

        for _ in range(cycles):
            self.transition()

    def show_history(self):

        print("State History:")

        for state in self.history:
            print(state)
```

```python
machine = SmartTrafficLight()

machine.run(10)

history = machine.history

plt.figure(figsize=(10,5))

plt.plot(history, marker='o')

plt.title("Traffic Light State Transitions")

plt.xlabel("Step")
plt.ylabel("State")

plt.grid(True)

plt.show()
```

```python
class StateMachine:

    def __init__(self, initial_state):

        self.state = initial_state
        self.transitions = {}

    def add_transition(self, state, event, next_state):

        self.transitions[(state, event)] = next_state

    def trigger(self, event):

        key = (self.state, event)

        if key in self.transitions:

            self.state = self.transitions[key]

            print("Transitioned to:", self.state)

        else:
            print("Invalid Event")
```

```python
sm = StateMachine("LOCKED")

sm.add_transition("LOCKED", "coin", "UNLOCKED")
sm.add_transition("UNLOCKED", "push", "LOCKED")

sm.trigger("coin")
sm.trigger("push")
```
