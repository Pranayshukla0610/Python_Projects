```python
#1. Prime Numbers Using the Sieve of Eratosthenes

def sieve_of_eratosthenes(n):
  if n < 2:
    return []

  sieve = [True] * (n + 1)
  sieve[0] = sieve[1] = False

  for i in range(2, int(n ** 0.5) + 1):
    if sieve[i]:
      for multiple in range(i*1, n+1, i):
        sieve[multiple] = False

  return [i for i, is_prime in enumerate(sieve) if is_prime]

  print(sieve_of_eratosthenes(20))
```

```python
#Sum of Digits of a Number

def sum_digits(n):
  n = abs(n)
  total = 0

  while n > 0:
    digit = n % 10
    total += digit
    n //= 10
  return total
print(sum_digits(12345))
```

```python
#Reverse a Number

def reverse_number(n):
  sign = -1 if n < 0 else 1
  n = abs(n)
  reverse_num = 0

  while n > 0:
    reverse_num = reverse_num * 10 + n % 10
    n //= 10
  return sign * reverse_num

print(reverse_number(1234))
```

```python
#Check for a Prime Number

def is_prime(n):
  if n <= 1:
    return []

  for i in range(2, int(n ** 0.5)+1):
    if n % i == 0:
      return False
    return True

print(is_prime(10))
```

```python
#Generate Fibonacci Series

def fibonacci_series(n):
  a, b = 0, 1
  series = []

  for _ in range(n):
    series.append(a)
    a, b = b, a+b

  return series

print(fibonacci_series(10))
```

```python
#Multiplication Table Generator

def multiplication_tables(n):
  for i in range(1, n+1):
    print(f"\nMultiplication Table  of {i}")
    for j in range(1,11):
      print(f"{i} * {j} = {i*j}")

multiplication_tables(5)
```

```python
#Count Vowels and Consonants in a String

def count_vowels_consonants(text):
  vowels = "aeiouAEIOU"
  vowel_count = 0
  consonant_count = 0

  for char in text:
    if char.isalpha():
      if char in vowels:
        vowel_count += 1
      else:
        consonant_count += 1

  return vowel_count, consonant_count

vowels, consonants = count_vowels_consonants("Hello World")
print(f"Vowels: {vowels}, Consonants: {consonants}")
```
