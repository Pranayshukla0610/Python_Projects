```python
import timeit
import tracemalloc
import matplotlib.pyplot as plt
import sys
```

```python
#Factorial Using Standard Recursion
def factorial(n):

  if n == 0:
    return 1

  return n * factorial(n - 1)
```

```python
#Tail Recursion Example

def tail_factorial(n, accumulator=1):

  if n == 0:
    return accumulator

  return tail_factorial(
      n - 1,
      n * accumulator
  )
```

```python
def fibonacci_standard(n):
  if n <= 1:
    return n

  return fibonacci_standard(n - 1) + fibonacci_standard(n - 2)
```

```python
def factorial_tail(n, accumulator=1):
  if n == 0:
    return accumulator

  return fibonacci_tail(
      n - 1,
      n * accumulator
  )
```

```python
def fibonacci_tail(n, a=0, b=1):

    if n == 0:
        return a

    return fibonacci_tail(
        n - 1,
        b,
        a + b
    )
```

```python
def factorial_standard(n):

    if n == 0:
        return 1

    return n * factorial_standard(n - 1)
```

```python
print("Standard Factorial:", factorial_standard(5))

print("Tail Factorial:", factorial_tail(5))

print("Standard Fibonacci:", fibonacci_standard(10))

print("Tail Fibonacci:", fibonacci_tail(10))
```

```python
standard_time = timeit.timeit(
    lambda: factorial_standard(500),
    number=100
)
```

```python
tail_time = timeit.timeit(
    lambda: factorial_tail(500),
    number=100
)
```

```python
print("Standard Recursion Time:", standard_time)

print("Tail Recursion Time:", tail_time)
```

```python
#Standard Recursion Memory

tracemalloc.start()

factorial_standard(300)

current, peak = tracemalloc.get_traced_memory()

print("Standard Peak Memory:", peak)

tracemalloc.stop()
```

```python
tracemalloc.start()

factorial_tail(300)

current, peak = tracemalloc.get_traced_memory()

print("Tail Peak Memory:", peak)

tracemalloc.stop()
```

```python
methods = [
    "Standard",
    "Tail"
]

times = [
    standard_time,
    tail_time
]

plt.figure(figsize=(8,5))

plt.bar(methods, times)

plt.xlabel("Recursion Type")

plt.ylabel("Execution Time")

plt.title("Standard vs Tail Recursion")

plt.show()
```
