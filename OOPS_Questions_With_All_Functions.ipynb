```python
#Employee Management System (Multi-Level Inheritence)

class Employee:
  def __init__(self,name,salary):
    self.name = name
    self._salary = salary


  def calculate_bonus(self):
    return self._salary  * 0.1

  def __str__(self):
    return f"{self.name} earns {self._salary}"


class Developer(Employee):
  def calculate_bonus(self):
    return self._salary * 0.2


class Manager(Employee):
  def calculate_bonus(self):
    return self._salary * 0.3


emps = [Developer("Pranay",50000), Manager("Rahul",80000)]

for e in emps:
  print(e, "Bonus:",e.calculate_bonus())
```

```python
#Library System with iterator

class Library:
    def __init__(self, name):
        self.name = name
        self.__books = []

    def add_book(self, book):
        self.__books.append(book)

    def __iter__(self):
        self.i = 0
        return self

    def __next__(self):
        if self.i >= len(self.__books):
            raise StopIteration
        book = self.__books[self.i]
        self.i += 1
        return book

    def __str__(self):
        return f"Library: {self.name}, Books: {len(self.__books)}"


# Usage
lib = Library("City Library")
lib.add_book("Python")
lib.add_book("ML")

for book in lib:
    print(book)
```

```python
#Ride Booking System (Uber-like Design)

from abc import ABC, abstractmethod

class Ride(ABC):
  def __init__(self, distance):
    self.distance = distance

  @abstractmethod
  def calculate_fare(self):
    pass

class BikeRide(Ride):
  def calculate_fare(self):
    return self.distance * 5

class CarRide(Ride):
  def calculate_fare(self):
    return self.distance * 10

class RideFactory:
  @staticmethod
  def get_ride(ride_type, distance):
    if ride_type == 'bike':
      return BikeRide(distance)
    elif ride_type == 'car':
      return CarRide(distance)

ride = RideFactory.get_ride("car", 10)
print("Fare:", ride.calculate_fare())
```
