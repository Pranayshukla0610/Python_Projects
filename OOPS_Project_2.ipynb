```python
#🏥 1. Hospital Management System
#Concepts:
#Inheritance → Patient, Doctor
#Encapsulation → patient records
#Polymorphism → treatment behavior
#Composition → Hospital contains departments

from abc import ABC, abstractmethod

class Person:
  def __init__(self,name):
    self.name = name

class Patient(Person):
  def __init__(self,name,illness):
    super().__init__(name)
    self.__illness = illness

  def get_illness(self):
    return self.__illness

class Doctor(Person):
  def treat(self,patient):
    print(f"Doctor treating {patient.name} for {patient.get_illness()}")

class Hospital:
  def __init__(self):
    self.patients = []
    self.doctors = []

  def add_patient(self,patient):
    self.patients.append(patient)

  def add_doctor(self,doctor):
    self.doctors.append(doctor)

  def assign_treatment(self):
    for doctor, patient in zip(self.doctors, self.patients):
      doctor.treat(patient)


h = Hospital()
h.add_patient(Patient("A", "Fever"))
h.add_doctor(Doctor("Dr. B"))

h.assign_treatment()
```

```python
#🚗 2. Vehicle Rental System
#Concepts:
#Inheritance → Vehicle types
#Encapsulation → rental pricing
#Polymorphism → pricing per vehicle
#Composition → Rental system manages vehicles

from abc import ABC, abstractmethod

class Vehicle(ABC):
    def __init__(self, name, base_price):
        self.name = name
        self.__base_price = base_price  # encapsulation

    @abstractmethod
    def rental_price(self, days):
        pass

    def get_base_price(self):
        return self.__base_price


class Car(Vehicle):
    def rental_price(self, days):
        return self.get_base_price() * days + 500


class Bike(Vehicle):
    def rental_price(self, days):
        return self.get_base_price() * days + 100


class RentalSystem:
    def __init__(self):
        self.vehicles = []

    def add_vehicle(self, vehicle):
        self.vehicles.append(vehicle)

    def calculate_rent(self, vehicle, days):
        return vehicle.rental_price(days)


car = Car("Sedan", 1000)
bike = Bike("Yamaha", 300)

system = RentalSystem()
print(system.calculate_rent(car, 3))
print(system.calculate_rent(bike, 3))
```
