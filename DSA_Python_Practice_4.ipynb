```python
#Find Uncommon Elements in Two Arrays

arr1 = [1,2,3,4]
arr2 = [3,4,5,6]

result = []

for num in arr1:
  if num not in arr2:
    result.append(num)

for num in arr2:
  if num not in arr1:
    result.append(num)

print(result)
```

```python
#Find Common Elements in Two Arrays

arr1 = [1,2,3,4]
arr2 = [3,4,5,6]

common = []

for num in arr1:
  if num in arr2:
    common.append(num)

print(common)
```

```python
#Check if Two Arrays are Equal After Sorting

arr1 = [30,10,20]
arr2 = [20,30,10]

if sorted(arr1) == sorted(arr2):
  print("Arrays are equal")
else:
  print("Arrays are not equal")
```

```python
#Check if Two Arrays are Identical

arr1= [10,20,30]
arr2= [30,20,10]

if arr1 == arr2:
  print("Arrays are identical")
else:
  print("Arrays are not identical")
```

```python
#Find the First Repeating Element

arr = [1,2,3,4,2,5]
seen = []

for num in arr:
  if num in seen:
    print(num)
    break

  seen.append(num)
```

```python
#Check Whether an Array is Sorted in Descending Order

arr = [50,40,30,20,10]


if arr == sorted(arr, reverse=True):
  print("Array is sorted in descending order")
else:
  print("Array is not sorted in descending order")
```

```python
#Find the Element with Maximum Frequency

arr = [1,2,2,3,3,3,4]

frequency = {}

for num in arr:
    frequency[num] = frequency.get(num,0)+1

maximum = max(frequency,key=frequency.get)

print(maximum)
```
