```python
#Rearrange Positives and Negatives Alternately

arr = [1, -2, 3, -4, 5, -6]

positive = []
negative = []

for num in arr:
  if num % 2 == 0:
    positive.append(num)
  else:
    negative.append(num)

result = []

i = 0

while i < len(positive) and i < len(negative):
  result.append(positive[i])
  result.append(negative[i])

  i += 1

if i < len(positive):
  result.extend(positive[i:])

if i < len(negative):
  result.extend(negative[i:])

print(result)
```

```python
#Left Rotate an Array by One Position

arr = [1, 2, 3, 4, 5]

first = arr.pop(0)

arr.append(first)

print(arr)
```

```python
#Right Rotate an Array by One Position

arr = [1, 2, 3, 4, 5]

right = arr.pop()

arr.insert(0,right)

print(arr)
```

```python
#Swap Adjacent Elements

arr = [1,2,3,4,5,6]

for i in range(0, len(arr)-1, 2):
  arr[i], arr[i+1] = arr[i+1], arr[i]

print(arr)
```

```python
#Perform Linear Search

arr = [10, 20, 30, 40, 50]

element = int(input("Enter an element to search:"))

for i in range(len(arr)):
  if arr[i] == element:
    print("Element found at:",i)
    break

else:
  print("Element not found")
```

```python
#Find Duplicate Elements

arr = [1,2,3,4,5]

duplicate = []

for num in arr:
  if arr.count(num) > 1 and num not in duplicate:
    duplicate.append(num)

print(duplicate)
```

```python
#Find Unique Elements

arr = [1, 2, 2, 3, 3, 4, 5]

for num in arr:
  if arr.count(num) == 1:
    print(num)
```

```python
#Remove Duplicate Elements

arr = [1,2,3,2,4,1,5]

result = []

for num in arr:
  if num not in result:
    result.append(num)

print(result)
```

```python
#Count Duplicate Elements

arr = [1,2,2,3,3,4,5]

count = 0

for num in arr:
  if arr.count(num) > 1:
    count += 1

print(count)
```

```python
#Find the Frequency of Every Element

arr = [1,2,2,3,3,3,4]

freq = {}

for num in arr:
  if num in freq:
    freq[num] += 1
  else:
    freq[num] = 1

print(freq)
```
