```python
#Find Uncommon Elements in Two Arrays

arr1 = [10, 20, 30, 40]
arr2 = [20, 30, 50, 60]

uncommon = []

for i in arr1:
  if i not in arr2:
    uncommon.append(i)

for j in arr2:
  if j not in arr1:
    uncommon.append(j)

print(uncommon)
```

```python
#Find Common Elements in Two Arrays

arr1 = [10, 20, 30, 40]
arr2 = [20, 30, 50, 60]

common = []

for i in arr1:
  if i in arr2:
    common.append(i)

print(common)
```

```python
#Check Whether an Array is Sorted in Descending Order

arr = [50, 40, 30, 20, 10]

descending = True

for i in range(len(arr)-1):
  if arr[i] < arr[i+1]:
    descending = False
    break

print(descending)
```

```python
#Check Whether an Array is Sorted

arr = [10, 20, 30, 40, 50]

sorted = True

for i in range(len(arr)-1):
  if arr[i] > arr[i+1]:
    sorted = False
    break

print(sorted)
```

```python
#Find the Element with Maximum Frequency

arr = [10, 20, 10, 30, 20, 20]

max_count = 0
element = 0

for i in arr:
  if arr.count(i) > max_count:
    max_count = arr.count(i)
    element = i

print(element)
```

```python
#Find the Element with Minimum Frequency

arr = [10, 20, 10, 30, 20, 20, 40]

min_count = len(arr)
element = None

for i in arr:
    if arr.count(i) < min_count:
        min_count = arr.count(i)
        element = i

print(element)
```

```python
#Print Only Distinct Elements

arr = [10, 20, 10, 30, 40, 20, 50]


for i in arr:
  if arr.count(i) == 1:
    print(i)
```

```python
#Count Duplicate Elements

arr = [10, 20, 10, 30, 20, 40]

count = 0
visited = []

for i in arr:
  if i not in visited and arr.count(i) > 1:
    count += 1
    visited.append(i)

print(count)
```

```python
#Find the Missing Number from 1 to N

arr = [1, 2, 3, 5]

n = 5

expected_sum = n * (n + 1)//2

actual_sum = 0

for i in arr:
  actual_sum += 1

print("Missing Number:",expected_sum - actual_sum)
```
