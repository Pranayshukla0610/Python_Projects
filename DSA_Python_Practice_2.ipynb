```python
#Insert an Element at the Beginning

arr = [20,30,50]

element = int(input("Enter a number:"))

arr.insert(0,element)

print(arr)
```

```python
#Insert an Element at the End

arr = [20,30,60]

element = int(input("Enter a number:"))

arr.insert(len(arr),element)

print(arr)
```

```python
#Insert an Element at the End

arr = [20,50,80]

element = int(input("Enter a number:"))

arr.append(element)

print(arr)
```

```python
#Insert an Element at a Given Position

arr = [10, 20, 40, 50]

element = int(input("Enter a number:"))
position = int(input("Enter a position:"))

arr.insert(position, element)

print(arr)
```

```python
#Delete the First Element

arr = [10, 20, 40, 50]

arr.pop(0)

print(arr)
```

```python
#Delete the Last Element

arr = [10, 20, 40, 50]

arr.pop()

print(arr)
```

```python
#Delete all occurrences of a value.

arr = [10, 20, 40, 60]

element = int(input("Enter a number:"))

for i in arr:
  if i == element:
    arr.remove(i)

print(arr)
```

```python
#Move all zeros at the beginning.

arr = [0, 1, 0, 2, 3, 6]

result = []

for num in arr:
  if num != 0:
    result.append(num)

  else:
    result.insert(0,num)

print(result)
```

```python
#Move all zeros to the end

arr = [0, 1, 0, 2, 3, 6]

result = []

for num in arr:
  if num != 0:
    result.append(num)

while len(result) < len(arr):
  result.append(0)

print(result)
```

```python
#Move All Negative Numbers to One Side

arr = [2, -1, -3, 6, -7, 8, 10]

result = []

for num in arr:
  if num < 0:
    result.append(num)

for num in arr:
  if num >= 0:
    result.append(num)

print(result)
```

```python
#Separate Even and Odd Numbers

arr = [4, 7, 8, 10, 15, 13, 6, 22, 21, 7]

result = []

for num in arr:
  if num % 2 == 0:
    result.append(num)

for num in arr:
  if num % 2 == 1:
    result.append(num)

print(result)
```

```python
#Positives and Negatives Alternately

arr = [1, -2, 3, -4, 5, -6]

pos_num = []
neg_num = []

for num in arr:
  if num >= 0:
    pos_num.append(num)
  else:
    neg_num.append(num)

print("Positive Numbers:",pos_num)
print("Negative Numbers:",neg_num)
```
