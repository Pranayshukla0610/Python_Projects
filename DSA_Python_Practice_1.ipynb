```python
#Find the Longest Common Prefix

#Input:
#["flower", "flow", "flight"]

#Output:
#fl

strings = ['flower','flow','flight']

prefix = strings[0]

for word in strings[1:]:
  while not word.startswith(prefix):
    prefix = prefix[:-1]


print(prefix)
```

```python
#Find the maximum element in an array

arr = [25, 15, 60, 10, 90, 45]

max_arr = arr[0]

for i in arr:
  if i > max_arr:
    max_arr = i

print(max_arr)
```

```python
#Count even numbers in an array

arr = [10, 15, 25, 30, 33, 47, 48, 51]

count = 0

for i in arr:
  if i % 2 == 0:
    count += 1

print(count)
```

```python
#Count odd numbers in an array

arr = [15, 17, 18, 21, 22, 23, 25, 27]

count = 0

for i in arr:
  if i % 2 != 0:
    count += 1

print(count)
```

```python
#Find the largest and smallest element together

arr = [25, 15, 60, 10, 90, 45]

largest = arr[0]
smallest = arr[0]

for i in arr:
  if i > largest:
    largest = i
  elif i < smallest:
    smallest = i

print(largest)
print(smallest)
```

```python
#Find the second largest element

arr = [25, 15, 60, 10, 90, 45]

largest = arr[0]
second_largest = arr[0]

for i in arr:
  if i > largest:
    largest = i
  elif second_largest < i < largest:
    second_largest = i

print(second_largest)
```

```python
#Find the second smallest element5

arr = [25, 15, 60, 10, 90, 45]

smallest = float('inf')
second_smallest = float('inf')

for i in arr:

    if i < smallest:
        second_smallest = smallest
        smallest = i

    elif smallest < i < second_smallest:
        second_smallest = i

print("Second Smallest =", second_smallest)
```
