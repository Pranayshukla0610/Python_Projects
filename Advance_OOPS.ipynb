```python
#Create a CLI program where users can add students and display them

class student:
  def __init__(self,name,age):
    self.name = name
    self.age = age

students = []

while True:
  print("\n1 Add Students")
  print("2 Show Students")
  print("3 Exit")

  choice = input("Enter your choice:")

  if choice == "1":
    name = input("Enter student name:")
    age = int(input("Enter student age:"))

    student = student(name,age)
    students.append(student)

  elif choice == "2":
    for s in students:
      print(s.name,s,age)

  elif choice == "3":
    break
```

```python
#Create product inventory system

class product:
  def __init__(self,name,price):
    self.name=name
    self.price=price

products = []

while True:
  print("\n1 Add Products")
  print("2 Show Products")
  print("3 Exit")

  choice = input("Enter your choice:")

  if choice == "1":
    name = input("Enter product name:")
    price = float(input("Enter product price:"))

    product = product(name,price)
    products.append(product)

  elif choice == "2":
    for p in products:
      print(p.name,p.price)

  elif choice == "3":
    break
```

```python
#Bank Accounting System (Encapsulation)

class BankAccount:
  def __init__(self,owner):
    self.owner = owner
    self.__balance = 0

  def deposit(self,amount):
    self.__balance += amount

  def withdraw(self,amount):
    if self.__balance >= amount:
      self.__balance -= amount

  def show_balance(self):
    return self.__balance

account = BankAccount('Rahul')

while True:
  print("\n1 Deposit")
  print("2 Withdraw")
  print("3 Balance")
  print("4 Exit")

  c = input()

  if c == "1":
    account.deposit(int(input("Amount:")))

  elif c == "2":
    account.withdraw(int(input("Amount")))

  elif c == "3":
    print(account.show_balance())

  else:
    break
```

```python
#Library Stores Books

class Book:
  def __init__(self,title):
    self.title = title


class Library:
  def __init__(self):
    self.books = []

  def add_book(self,book):
    self.books.append(book)

  def show_book(self):
    for b in self.books:
      print(b.title)

lib = Library()

lib.add_book(Book("Python"))
lib.add_book(Book("SQL"))

lib.show_book()
```

```python
#Multiple Inheritence
#Manager inherits

class Person:
  def __init__(self,name):
    self.name = name

class Employee:
  def __init__(self,salary):
    self.salary = salary

class Manager(Person, Employee):
  def __init__(self,name,salary):
    Person.__init__(self,name)
    Employee.__init__(self,salary)

m = Manager("Rahul",60000)
print(m.name,m.salary)
```

```python
#Library Management System
class Book:
  def __init__(self,title):
    self.title = title
    self.is_borrowed = False

class Library:
  def __init__(self):
    self.books = []

  def add_book(self,book):
    self.books.append(book)

  def borrow_book(self,title):
    for b in self.books:
      if b.title == title and not b.is_borrowed:
        b.is_borrowed = True
        return
    print("Book is not available")

  def return_book(self,title):
    for b in self.books:
      if b.title == title and b.is_borrowed:
        b.is_borrowed = False
        print("Book returned")
        return

  def show_books(self):
    for b in self.books:
      status='Borrowed' if b.is_borrowed else 'Available'
      print(b.title, "-", status)

lib = Library()
lib.add_book(Book("Python"))
lib.add_book(Book("Machine Learning"))

lib.borrow_book("Python")
lib.show_books()
```

```python
#Banking System
class Account:
  def __init__(self,owner,balance=0):
    self.owner = owner
    self.__balance = balance

  def deposit(self,amount):
    self.__balance += amount

  def withdraw(self,amount):
    if amount <= self.__balance:
      self.__balance -= amount
    else:
      print("Insufficient fund")

  def get_balance(self):
    return self.__balance


class Bank:
  def __init__(self):
    self.accounts = {}

  def create_account(self,owner):
    acc = Account(owner)
    self.accounts[owner] = acc

  def get_account(self,owner):
    return self.accounts.get(owner)


bank = Bank()
bank.create_account("Alice")

acc = bank.get_account("Alice")
acc.deposit(5000)
acc.withdraw(2000)

print(acc.get_balance())
```
